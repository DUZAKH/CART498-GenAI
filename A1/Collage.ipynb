```python
from diffusers import StableDiffusionPipeline
import torch
model_id = "prompthero/openjourney"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
prompt = "flatbed scan of old surreal paper collage, dull frayed cardboard background, small flat magazine cutout of full body painted soldier at bottom rough paper edges, scan noise and static lines, soft glowing star cutouts, xerox zine aesthetic"
image = pipe(prompt).images[0]
image.save("./collage.png")

```
