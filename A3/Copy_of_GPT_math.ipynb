```python
# Access your secret keys via
from google.colab import userdata
# The name of your secret must match `OPENAI_API_KEY`
OPENAI_API_KEY = userdata.get('me_API')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))
```

```python
print('--- AI-driven Math Calculator ---')
print('Type your math problem or type "quit" to exit.')

def calculate_iterated_square(n, i):
     if i < 0:
         raise ValueError("Number of iterations 'i' must be non-negative.")
     if i == 0:
         return n
     result = n * n
     for _ in range(1, i):
         result *= result
     return result

error_count = 0
prev_error =0;
# self_deprecating_messages = [
#     "Wow, I messed up again. Math is hard, okay?",
#     "Seriously? Another one? My circuits are getting tangled.",
#     "This is getting embarrassing. Maybe I should stick to writing poems.",
#     "I'm starting to think 'AI-driven' might be a bit ambitious for me today.",
#     "Okay, I'm officially defeated. Someone get me a rubber duck."
# ]

while True:
    try:
      n_input = input("Enter the base number (n): ")
      if n_input.lower() == 'quit':
            print('Exiting calculator. Goodbye!')
            break
      n = float(n_input)
      i_input = input("Enter the number of iterations (i): ")
      if i_input.lower() == 'quit':
            print('Exiting calculator. Goodbye!')
            break
      i = int(i_input)
      user_input = f"{n}^(2^{i})"
      correct_res = calculate_iterated_square(n, i)
      response = client.chat.completions.create(
            model="gpt-4.1-nano",
            temperature=0.9,  # Low temperature for deterministic output
            top_p=0.0,        # Low top_p to focus on most probable tokens
            messages=[
                {
                    "role": "user",
                    "content": f"Solve the following math problem. Provide only the numerical answer without any additional text or explanation: {user_input}"

                  # ,
                    # "role": "developer",
                    # "content": f"you are a very critical person. you have a personality that gets frustrated or defeated over multiple errors. If the result is not:{correct_res}, generate a humorous self-deprecating response. "
                }
            ]
        )
      result = response.choices[0].message.content.strip()
      print(f"Result from AI: {result}")
      print(f"Correct Result (n^(2^i)): {correct_res}")

      # # Self-deprecation mechanism
      if float(result) != correct_res:
         error_count=error_count+1
         response_dep = client.chat.completions.create(
            model="gpt-4.1-nano",
            temperature=0.9,  # Low temperature for deterministic output
            top_p=0.0,        # Low top_p to focus on most probable tokens
            messages=[
                {

                    "role": "developer",
                     "content": f"you are a very critical person. you have a personality that gets frustrated or defeated over multiple errors, generate a humorous self-deprecating response. If the {error_count} is greater than the {prev_error} then become more deprecating"
                }
            ]
        )
         result_dep = response_dep.choices[0].message.content.strip()
         print(f"Result from AI dep: {result_dep}")
         prev_error = error_count

      # #     print("My calculation is correct! Feeling smart today. 😉")
      # #     error_count = 0 # Reset error count on success
      # else:
      #     print("My calculation is incorrect. Uh oh...")
      #     error_count += 1
      #     message_index = min(error_count - 1, len(self_deprecating_messages) - 1)
      #     print(self_deprecating_messages[message_index])

    except ValueError:
        print("Invalid input. Please enter valid numbers for n and i.")
    except Exception as e:
        print(f"An error occurred: {e}")
        print("Please try again or enter 'quit'.")
```
